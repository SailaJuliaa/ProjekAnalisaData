## Visualization & Explanatory Analysis

### Bagaimana tren jumlah peminjaman sepeda harian dalam satu tahun

```python
plt.figure(figsize=(20, 10))
sns.lineplot(data=df_day, x="dteday", y="cnt", label="Total Rentals")
plt.title("Jumlah peminjaman sepeda harian dalam satu tahun")
plt.xlabel("Tanggal")
plt.ylabel("Number rentals")
plt.xticks(rotation=45)
plt.show()
```

###Bulan apa yang memiliki jumlah peminjaman tertinggi dan terendah

```python
df_day["cnt"] = pd.to_numeric(df_day["cnt"], errors="coerce")
df_hour["cnt"] = pd.to_numeric(df_hour["cnt"], errors="coerce")
```

```python
monthly_rentals = df_day.groupby("mnth", as_index=False).agg({"cnt": "sum"})
```

```python
max_month = monthly_rentals.loc[monthly_rentals["cnt"].idxmax(), "mnth"]
max_rentals = monthly_rentals["cnt"].max()
min_month = monthly_rentals.loc[monthly_rentals['cnt'].idxmin(), "mnth"]
min_rentals = monthly_rentals["cnt"].min()

print(f"Bulan yang memiliki peminjaman tertinggi: {max_month} ({max_rentals:,} rentals)")
print(f"Bulan yang memiliki peminjaman terendah : {min_month} ({min_rentals:,} rentals)")
```

```python
df_day["mnth"] = df_day["dteday"].dt.month
```

```python
monthly_rentals = df_day.groupby("mnth")["cnt"].sum()
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_rentals.index, y=monthly_rentals.values, palette="flare")
plt.title("Total rental sepeda per bulan")
plt.xlabel("Bulan")
plt.ylabel("Total Rentals")
plt.xticks(ticks=range(0, 12), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

**Insight:**
- Pada visualisasi data menganai bagaimana tren jumlah peminjaman sepeda harian dalam satu tahun yaitu : Mengalami peningkatan pada tahun 2011 awal dan penurunan di 2011 akhir sampai 2012 awal, kemudian mengalami peningkatkan pesat di 2012 pada bualan april sampai dengan november dan mengalami penurunan kembali hingga 2013
- Dalam arti lain peningkatan peminjaman bisa terjadi berdasarkan pola musiman terutaman pada musim panas
- Bulan dengan jumlah peminjaman tertinggi dan terendah terdapat pada bulan agustus dan jumlah peminjaman terendah di bulan januari

## Analisis Lanjutan (Opsional)

**Analisis pola musiman**
Dapat dilihat dari jumlah peningkatan peminjaman sepeda bahwa pada musim panas cenderung lebih banyak menghasilkan peminjaman dibanding musim lain itu berarti adanya keterlibatan antara temp dan atemp, maka strategi yang bisa dilakukan kedepannya untuk menghasilkan peminjaman lebih banyak lagi adalah
*menawarkan diskon dan promo pada musim dingin untuk menarik pelanggan dan menambah jumlah sepeda pada musim panas agar menghindari lonjakan peminjaman*

**Analisis pola harian**
Pada data hour bisa dilihat jam jam dengan peminjaman yang tinggi pada jam berangkat dan pulang aktivitas yaitu pagi hingga sore, maka strategi yang bisa dilakukan kedepannya yaitu *penyesuain harga pada jam dengan puncak peminjaman yang tinggi*
